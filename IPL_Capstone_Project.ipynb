# Using Pure Python

```python
match_wins = {}

for team in df['match_winner']:
    if team in match_wins:
        match_wins[team] += 1
    else:
        match_wins[team] = 1

print(match_wins)
```

# Using Pandas

```python
match_wins = df['match_winner'].value_counts()

print(match_wins)
```

# Using Matplotlib

```python
import matplotlib.pyplot as plt

match_wins = df['match_winner'].value_counts()

plt.figure(figsize=(12,5))
plt.bar(match_wins.index, match_wins.values)

plt.title("Most Match Wins by Team")
plt.xlabel("Teams")
plt.ylabel("Wins")

plt.xticks(rotation=45)
plt.show()
```

# Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

match_wins = df['match_winner'].value_counts()

plt.figure(figsize=(12,5))

sns.barplot(
    x=match_wins.index,
    y=match_wins.values
)

plt.title("Most Match Wins by Team")
plt.xlabel("Teams")
plt.ylabel("Wins")

plt.xticks(rotation=45)
plt.show()
```

### Custom Questions & Insights

Q3: Which bowler had the best bowling figures?

```python
df.head()
```

```python
df[df['highest_wickets'] == df['highest_wickets'].max()][['best_bowling','best_bowling_figure']]
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[0])
df['highest_wickets_with_most_economy'] = df['best_bowling_figure'].apply(lambda x :x.split('--')[1])
df['highest_wickets'] = df['highest_wickets'].astype(int)
df['highest_wickets_with_most_economy'] = df['highest_wickets_with_most_economy'].astype(int)

best_figures = df.sort_values(
    by=['highest_wickets', 'highest_wickets_with_most_economy'],
    ascending=[False, True]
)

print(best_figures[
    ['best_bowling',
     'highest_wickets',
     'highest_wickets_with_most_economy']
].head(1))
```
